```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Washington DC center coords
LAT_CENTER = 38.9072
LON_CENTER = -77.0369
SPREAD = 0.02  # ~2km radius

np.random.seed(42)

# A few fixed hotspot locations (shared across targets)
hotspots = [
    (38.9072, -77.0369),  # Downtown DC
    (38.9145, -77.0298),  # Near Capitol
    (38.8977, -77.0365),  # Near Mall
]

records = []
base_date = datetime(2026, 3, 1)

for day in range(10):        # 10 days
    dt_base = base_date + timedelta(days=day)
    for _ in range(10):      # 10 sightings per day
        target = np.random.randint(1, 7)

        # 40% chance of landing on a hotspot
        if np.random.random() < 0.4:
            lat, lon = hotspots[np.random.randint(len(hotspots))]
            lat += np.random.uniform(-0.001, 0.001)
            lon += np.random.uniform(-0.001, 0.001)
        else:
            lat = LAT_CENTER + np.random.uniform(-SPREAD, SPREAD)
            lon = LON_CENTER + np.random.uniform(-SPREAD, SPREAD)

        hour   = np.random.randint(6, 22)
        minute = np.random.randint(0, 60)
        last_seen = dt_base.replace(hour=hour, minute=minute)

        records.append({
            "target":    target,
            "last_seen": last_seen.strftime("%Y-%m-%d %H:%M"),
            "lat":       round(lat, 6),
            "lon":       round(lon, 6),
        })

df = pd.DataFrame(records)
df.to_csv("Data/sightings.csv", index=False)
```

```python
print(f"\nTotal records: {len(df)}")
print(df["target"].value_counts().sort_index())
df.sort_values("target")['target'].value_counts()
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Date range — last 90 days
dates = pd.date_range(end=pd.Timestamp.today(), periods=90, freq="D")

records = []
for dt in dates:
    # Simulate realistic ingest patterns:
    # - Weekdays higher than weekends
    # - Some random spikes and dips
    # - Gradual upward trend over time
    base     = 1000 + (dt.dayofyear * 2)           # slow growth trend
    weekday  = 1.0 if dt.weekday() < 5 else 0.4    # weekends ~40% of weekday
    noise    = np.random.normal(1.0, 0.15)          # ±15% random noise
    spike    = 3.0 if np.random.random() < 0.05 else 1.0  # 5% chance of spike

    count = max(0, int(base * weekday * noise * spike))

    records.append({
        "datetime":   dt,
        "date":       dt.date(),
        "day_of_week": dt.strftime("%A"),
        "is_weekend":  dt.weekday() >= 5,
        "data_count":  count,
    })

df = pd.DataFrame(records)
df.to_csv("Data/daily_ingest_data.csv", index=False)

print(df.head(20))
print(f"\nTotal records : {len(df)}")
print(f"Date range    : {df['date'].min()} → {df['date'].max()}")
print(f"Avg daily     : {df['data_count'].mean():.0f}")
print(f"Max daily     : {df['data_count'].max()}")
print(f"Min daily     : {df['data_count'].min()}")
```
